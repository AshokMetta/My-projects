```python
data=pd.read_csv("D:\Assignment (NUmpy,Visualisation, MOdels)\census-income.csv",na_values="?",skipinitialspace=True)
data
```

```python
#● Rename the last column as Annual Income
data=data.rename({"Unnamed: 14":"Annual-Income"},axis=1)
```

```python
data.dropna(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

for col in data.columns:
    data[col].dtype=="Object"
    la=LabelEncoder()
    data[col]=la.fit_transform(data[col])
```

```python
data.head()
```

```python
data
```

```python
X=data.iloc[:,:-1]
Y=data.iloc[:,-1]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=0,test_size=0.3)
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
ypred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_test,ypred)
```

```python
confusion_matrix(y_test,ypred)
```

```python
print(classification_report(y_test,ypred))
```

```python
# Checking using the StandardScaler
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
xx_train=scale.fit_transform(x_train)
xx_test=scale.transform(x_test)
```

```python
lr1=LogisticRegression()
lr1.fit(xx_train,y_train)
```

```python
y_pred=lr1.predict(xx_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)       #[[TN-FP],[FN-TP]]
```

```python
print(classification_report(y_test,y_pred))
```
