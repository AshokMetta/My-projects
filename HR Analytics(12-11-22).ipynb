```python
a = ['dept','numberOfProjects','timeSpent.company','workAccident','promotionInLast5years','salary']
```

```python
a
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig = plt.subplots(figsize = (20,35)) # I want to plot multiple plots all at once.
for i,j in enumerate(a):
# The enumerate() helps out in extracting the data along with its index value
# i --> index value   j --> data present at that particular index value
    plt.subplot(3,2,i+1)  # (3,2,1) (3,2,2) (3,3,3) .... (3,3,6)
    plt.subplots_adjust(hspace = 1.0,wspace = 0.2)
    sns.countplot(x = j, data = df, hue = 'left') # hue --> for categorical data
    plt.xticks(rotation=90)
plt.show()
```

```python
fig = plt.subplots(figsize = 20,25)
plt.subplot(3,2,1)
sns.countplot(x = 'dept', data = df, hue = 'left')

plt.subplot(3,2,2)
sns.countplot(x = 'number', data = df, hue = 'left')
```
