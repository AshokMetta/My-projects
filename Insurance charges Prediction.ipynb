```python
obj_dtypes=data.select_dtypes("object")
obj_dtypes.nunique()
```

```python
from sklearn.preprocessing import LabelEncoder
Label_enc=LabelEncoder()
for i in list(obj_dtypes.columns):
    obj_dtypes[i]=Label_enc.fit_transform(obj_dtypes[i])
obj_dtypes
```

```python
data.isnull().sum()
```

```python
data['age']=data['age'].fillna(data['age'].mean())
```

```python
data.isnull().sum()
```

```python
col_names=data.columns
data.info()
```

```python
for col in col_names:
    if data[col].dtype=="object":
        data[col]=data[col].fillna(data[col].mode())
    else:
        data[col]=data[col].fillna(data[col].mean())
```

```python
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
for col in col_names:
    if data[col].dtype=="float64" or data[col].dtype=="int64":
        plt.boxplot(data[col])
        plt.xlabel(col)
        plt.ylabel("count")
        plt.show()
```

```python
Q1= data['bmi'].quantile(0.25)
Q3= data['bmi'].quantile(0.75)
IQR=Q3-Q1
data=data[(data['bmi']>=Q1-1.5*IQR)&(data['bmi']>=Q3+1.5*IQR)]
```

```python
plt.boxplot(data['bmi'])
plt.xlabel('bmi')
plt.ylabel("count")
plt.show()
```

```python


Q1= data.past_consultations.quantile(0.25)
Q3= data.past_consultations.quantile(0.75)
IQR=Q3-Q1
data=data[(data["past_consultations"]>=Q1-1.5*IQR)&(data["past_consultations"]>=Q3+1.5*IQR)]

Q1= data.Hospital_expenditure.quantile(0.25)
Q3= data.Hospital_expenditure.quantile(0.75)
IQR=Q3-Q1
data=data[(data["Hospital_expenditure"]>=Q1-1.5*IQR)&(data["Hospital_expenditure"]>=Q3+1.5*IQR)]

Q1= data.Anual_Salary.quantile(0.25)
Q3= data.Anual_Salary.quantile(0.75)
IQR=Q3-Q1
data=data[(data["Anual_Salary"]>=Q1-1.5*IQR)&(data["Anual_Salary"]>=Q3+1.5*IQR)]

```

```python
import matplotlib.pyplot as plt
for col in col_names:
    if data[col].dtype!="object":
        plt.boxplot(data[col])
        plt.xlabel(col)
        plt.ylabel("count")
        plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
col_list=[]
for col in col_names:
    if data[col].dtype!="object" and col!="charges":
        col_list.append(col)
        
x=data[col_list]
vif_data=pd.DataFrame()
vif_data["column names"]=x.columns
vif_data["VIF Values"]=[variance_inflation_factor(x.values,i)for i in range(len(x.columns))]
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

col_list = [col for col in col_names if data[col].dtype != "object" and col != "charges"]
x = data[col_list]

vif_data = pd.DataFrame()
vif_data["column names"] = x.columns
vif_data["VIF Values"] = [variance_inflation_factor(x.values, i) for i in range(len(x.columns))]

print(vif_data)
```

```python
for i in list(data.columns):
    if data[i].dtype=="int":
        x=data.iloc[:,:-1]
y=pd.DataFrame(data.iloc[:,-1])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
ln=LinearRegression()
ln.fit(x_train,y_train)
pred_y=ln.predict(x_test)
pred_y
```

```python
from sklearn.metrics import mean_absolute_percentage_error
MAE=mean_absolute_percentage_error(pred_y,y_test)
MAE
```
