```python
import fitz  # PyMuPDF

def pdf_to_text(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""

    for page_num in range(doc.page_count):
        page = doc[page_num]
        text += page.get_text()

    doc.close()
    return text

```

```python
# Example usage:
pdf_path = "Digital Personal Data Protection Act (1).pdf"
text_content = pdf_to_text(pdf_path)
print(text_content)
```
