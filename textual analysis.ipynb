```python
import nltk

def count_syllables(sentence):
    syllable_count = 0
    tokens = nltk.word_tokenize(sentence.lower())
    tokens = [token for token in tokens if token not in string.punctuation]
    tokens = [token for token in tokens if token not in stopwords.words('english')]
    for word in tokens:
        if word[-2:].lower() in ('es', 'ed'):
            syllable_count -= 1
        syllable_count += textstat.syllable_count(word)
    return syllable_count
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
import string
from collections import Counter
from textstat import textstat


all_data=[]
sid = SentimentIntensityAnalyzer()

# Load the stop words from multiple external files
stopwords_files = ['StopWords_Names.txt', 'StopWords_Geographic.txt', 'StopWords_GenericLong.txt', 'StopWords_Generic.txt', 'StopWords_DatesandNumbers.txt', 'StopWords_Currencies.txt', 'StopWords_Auditor.txt']
stopwords_list = []
for file in stopwords_files:
    with open(file, 'r') as a:
        stopwords_list += [word.strip() for word in a]
file_data_list = []

for file in filepath:
    with open(file, 'r', encoding='ISO-8859-1') as f:
        text = f.read()
    sentences = nltk.sent_tokenize(text)
    
    pos_scores = []
    neg_scores = []
    polarity_scores = []
    subjective_scores = []
    word_counts = []
    complex_counts=[]
    complex_word_count = 0
    s=0
    sentences_length=[]
    word_lengths=[]
    num_pronouns=[]
    syllables_count=[]
    for sentence in sentences:
        
        syllable_count = 0
        # Tokenize the sentence and remove punctuation
        tokens = nltk.word_tokenize(sentence)
        tokens = [token.lower() for token in tokens if token not in string.punctuation]

        # Remove stop words
        tokens = [token for token in tokens if token not in stopwords.words('english')]
        tokens = [token for token in tokens if token not in stopwords_list]
        
        syllable_count = count_syllables(sentence)
        syllables_count.append(syllable_count)
        word_length = sum(len(token) for token in tokens)
        word_lengths.append(word_length)
        word_count=Counter(1 for token in tokens)
        total = sum(word_count.values())
        word_counts.append(total)
        for word in tokens:
            if textstat.syllable_count(word) >= 2:
                complex_word_count+=1
        
        personal_pronouns = ['I', 'we','my', 'us','ours']
        pronouns_count=sum( token.lower in personal_pronouns for token in tokens)
        num_pronouns.append(pronouns_count)
        pos_word_count= sum(1 for word in tokens if word in P_list)
        pos_scores.append(pos_word_count)
        neg_word_count = sum(1 for word in tokens if word in N_list)
        neg_scores.append(neg_word_count)
        s+=1
        sentence_lenght=len(sentence)
        sentences_length.append(sentence_lenght)
        
        
        complex_count = textstat.difficult_words(text)
        complex_counts.append(complex_count)
        
    
    Positive_score=sum(pos_scores)
    Negative_score=sum(neg_scores)
    total_word_count=sum(word_counts)
    Avg_sentence_length=sum(sentences_length)/s
    complex_percentage = (sum(complex_counts) / total_word_count) * 100
    fox_index = 0.4 * ((total_word_count / s ) + (complex_percentage))
    AVG_no_of_words_per_sentence = total_word_count/s
    average_word_length= sum(word_lengths)/total_word_count
    No_of_pronouns=sum(num_pronouns)
    polarity_scores = (Positive_score - Negative_score)/((Positive_score + Negative_score)+0.000001)
    subjective_score = (Positive_score + Negative_score)/(total_word_count + 0.000001)
    syllabul_count=sum(syllables_count)
    # Print the results
    print(f'URLid:{file}:')
    print(f'Positive score: {Positive_score}')
    print(f'Negative score: {Negative_score}')
    print(f'Polarity scores: {polarity_scores}')
    print(f'subjective score: {subjective_score}')
    print(f'Average sentence length:{Avg_sentence_length}')
    print(f'complex percentage:{complex_percentage}')
    print(f'Fox Indes:{fox_index}')
    print(f'Average Number of Words Per Sentence: {AVG_no_of_words_per_sentence}')
    print(f'Complex words:{complex_word_count}')
    print(f'Word count: {total_word_count}')
    print(f'Syllabul count:{syllabul_count}')
    print(f'Number of Pronouns:{No_of_pronouns}')
    print(f'Average word length:{average_word_length}')
    print(f'Number of sentences:{s}')
    print('-'*50)
    file_data = [Positive_score, Negative_score,polarity_scores, subjective_score,Avg_sentence_length,complex_percentage
                ,fox_index,AVG_no_of_words_per_sentence,complex_word_count,total_word_count,syllabul_count,No_of_pronouns,average_word_length]
    file_data_list.append(file_data)

# Append the list of file results to the overall data list

files_data=list(file_data_list)

```
