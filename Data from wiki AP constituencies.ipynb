```python
import pandas as pd
url="https://en.wikipedia.org/wiki/List_of_constituencies_of_the_Andhra_Pradesh_Legislative_Assembly"
tables=pd.read_html(url)
tables
```

```python
table=tables[0]
```

```python
table
```

```python
table["Constituencies"]=table["Constituencies"].str.split('+',1,expand=True)[0]
```

```python
table
```

```python
table.to_excel('Constituencies.xlsx', index=False)
```
